```python
import string
import time
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

```python
#Chat Treatment
```

```python
chat_words= ""
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)

    return " " .join(new_text)
```

```python
# Remove Stopword
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return ' '.join(x)
```

```python
remove_stopwords("one of the other reviewers has mentioned that ")
```

```python
# Remove Emoji
```

```python
import re

def remove_emoji(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map
        u"\U0001F1E0-\U0001F1FF"  # flags
        u"\U00002700-\U000027BF"  # dingbats
        u"\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
text = "Hello 😊🚀! This is a test 👍🔥"
clean_text = remove_emoji(text)

print(clean_text)
```

```python
# Stemming
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join(ps.stem(word) for word in text.split())
```

```python
sample = "walks walking walked walk"
stem_words(sample)
```

```python
# Lemmatization
```

```python
from nltk.stem import WordNetLemmatizer
import nltk

wordnet_lemmatizer = WordNetLemmatizer()

sentence = " He is going to shopping because purchasing some product which is very helpful for life"
punctuation = "):!.,;"

sentence_words = nltk.word_tokenize(sentence)

for word in sentence_words:
    if word in punctuation:
        sentence_words.remove(word)


sentence_words

print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos = 'v')))
```
