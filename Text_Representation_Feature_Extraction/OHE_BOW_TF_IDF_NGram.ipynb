```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'text': ['people watch movie', 'movie write movie', 'people write comment', 'movie write comment'],
    'output': [1, 1, 0, 0]
})
```

```python
df
```

```python
# Bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv =  CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
bow
```

```python
# vocablury
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
```

```python
cv.transform(['Movie write comment and people watch movie']).toarray()
```

```python
# N gram
#Unigram
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv =  CountVectorizer(ngram_range=(1,1))
```

```python

ugram = cv.fit_transform(df['text']).toarray()
ugram
```

```python
#Bigram
```

```python
cv =  CountVectorizer(ngram_range=(2,2))
bigram = cv.fit_transform(df['text']).toarray()
bigram
```

```python
print(cv.vocabulary_)
```

```python
# Trigram
```

```python
cv =  CountVectorizer(ngram_range=(3,3))
trigram = cv.fit_transform(df['text']).toarray()
trigram
```

```python
print(cv.vocabulary_)
```

```python
# TF-IDF
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
```

```python
print(tfidf.get_feature_names_out())
```
